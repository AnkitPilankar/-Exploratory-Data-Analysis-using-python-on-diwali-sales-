```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Diwali Sales Data.csv",encoding= 'unicode_escape')
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.rename(columns={'Marital_Status':'shaadi'})
```

```python
df.describe()
```

```python
ax= sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#state

sales_stats = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(15,5))
bar_plot=sns.barplot(x='State',y='Orders',data=sales_stats)
for bars in bar_plot.containers:
  bar_plot.bar_label(bars)
```

```python
# to check how much amount/sales from top 10 states
#for below graph it tell us that 'UP' has highest amount
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize=(15,5))
bar_plot=sns.barplot(x='State',y='Amount',data = sales_state)

```

```python
# total count of marital_status, 0 = married , 1 = single
ax=sns.countplot(data=df,x='Marital_Status')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# from below graph we get that married and single women spent more amount
marital_amount = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(x='Marital_Status',y='Amount',data = marital_amount,hue='Gender')
```

```python
df['Occupation'].value_counts()
```

```python
#here we have count top 5 occupation count
z=df['Occupation'].value_counts().head(5)
plt.figure(figsize=(15,5))

ax=sns.countplot(data=df,x='Occupation',order=z.index)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# here we group by occupation basis on amount we see that it sector has more purchasing amount
occupation_amount = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(15,5))
ax=sns.barplot(x='Occupation',y='Amount',data = occupation_amount)
```

```python
# product_category
# here from below graph we can say food category spent more on amount

product_amount = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize=(15,5))
ax=sns.barplot(x='Product_Category',y='Amount',data = product_amount)
```
